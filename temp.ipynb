```python
import pandas as pd
df = pd.read_csv('src/application_train.csv')
df
```

```python
unused_columns = ["NAME_TYPE_SUITE", "WEEKDAY_APPR_PROCESS_START", "HOUR_APPR_PROCESS_START",
                      "WALLSMATERIAL_MODE", "FLAG_MOBIL", "ORGANIZATION_TYPE"] + \
                     [col for col in df.columns if col.endswith('_MODE') or col.endswith('_MEDI')]
df.drop(columns=unused_columns, inplace=True)
```

```python
df.dropna(thresh=len(df.index)/2, axis=1, inplace=True)
df
```

```python
df["AGE"] = round((-1 * df["DAYS_BIRTH"]) / 365, 0)
df.drop(columns=['DAYS_BIRTH'], inplace=True)
df["YEARS_EMPLOYED"] = round((-1 * df["DAYS_EMPLOYED"]) / 365, 2)
df.drop(columns=['DAYS_EMPLOYED'], inplace=True)
df["YEARS REGISTERED"] = round((-1 * df["DAYS_REGISTRATION"]) / 365, 2)
df.drop(columns=['DAYS_REGISTRATION'], inplace=True)
df["YEARS_ID_PUBLISH"] = round((-1 * df["DAYS_ID_PUBLISH"]) / 365, 2)
df.drop(columns=['DAYS_ID_PUBLISH'], inplace=True)

df.head()
```

```python
cat_columns = df.select_dtypes(include=['object']).columns
numerics = df.select_dtypes(include=['number']).columns

df[cat_columns] = df[cat_columns].fillna(df[cat_columns].mode().iloc[0])
df[numerics] = df[numerics].fillna(df[numerics].mean())
```

```python
df
```

```python
#Make Sure all NaNs are gone
df.dropna(axis=0, inplace=True)
df
```

```python
# save to csv
df.to_csv('application.csv', index=False)
```
